```python
def extract_airbnb_data(urls_list):
    for url in urls_list:
        airbnb_url = str(url)
        response = requests.get(airbnb_url)

        soup = BeautifulSoup(response.content, 'html.parser')
        
        master_list = []

        
    for item in soup.find_all('div', {'itemprop': "itemListElement"}):
        data_dict = {}

        try:
            print(item.select('[itemprop="name"]')[0]['content'])
            data_dict['name'] = item.select('[itemprop="name"]')[0]['content']

            print(item.select('[itemprop="url"]')[0]['content'])
            data_dict['url'] = item.select('[itemprop="url"]')[0]['content']

            print(item.find_all('span', {'role': "img"})[0].text)
            data_dict['ratings'] = item.find_all('span', {'role': "img"})[0].text

            print(item.find_all('button', {'class': "_nobfedm"})[0].text)
            data_dict['total_money'] = item.find_all('button', {'class': "_nobfedm"})[0].text

            print(item.text.split("·")[1:])
            data_dict['rest'] = item.text.split("·")[1:]

            master_list.append(data_dict)

        except Exception as e:
            print('')

        print('_' * 40)
        
        
    df = pd.DataFrame(master_list)
    
    return df
```

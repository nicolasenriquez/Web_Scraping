```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Total amount of pages 50 from 1-50 --> 1000 books to scrape info

number_pages_to_scrape = 10
 
books_to_scrap_url = "http://books.toscrape.com/catalogue/category/books_1/page-1.html"
response = requests.get(books_to_scrap_url)

# All html
soup = BeautifulSoup(response.text, "html.parser")
```

```python
bookshelf = soup.find_all('li', {'class': "col-xs-6 col-sm-4 col-md-3 col-lg-3"})
```

```python
master_list = []

for books in bookshelf:
    data = {}
    
    print(books.h3.a['title'])
    book_title = books.h3.a['title']
    data['book_title'] = book_title
    
    print(books.find_all("p", {"class": "price_color"})[0].text.strip())
    book_price = books.find_all("p", {"class": "price_color"})[0].text
    data['book_price'] = book_price
    
    print(books.find_all("div", {"class": "image_container"})[0].a['href'].replace("../..",""))
    catalog = "https://books.toscrape.com/catalogue"
    book_url = catalog + books.find_all("div", {"class": "image_container"})[0].a['href'].replace("../..","")
    data['book_url'] = book_url
    
    print(books.p['class'][1])
    book_rating = books.p['class'][1]
    data['book_rating'] = book_rating
    
    master_list.append(data)
    
    print('-' * 50)
```

```python
master_list = []

start_clock = time.process_time_ns()

for page in range(1, 51):
    print(f"Page {page} beginning scraping...")
    
    books_url_scrap = f"http://books.toscrape.com/catalogue/category/books_1/page-{page}.html"
    response = requests.get(books_url_scrap)
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    bookshelf = soup.find_all('li', {'class': "col-xs-6 col-sm-4 col-md-3 col-lg-3"})
    
    for books in bookshelf:
        data = {}
        
        book_title = books.h3.a['title']
        data['book_title'] = book_title

        book_price = books.find_all("p", {"class": "price_color"})[0].text
        data['book_price'] = book_price

        catalog = "https://books.toscrape.com/catalogue"
        book_url = catalog + books.find_all("div", {"class": "image_container"})[0].a['href'].replace("../..","")
        data['book_url'] = book_url

        book_rating = books.p['class'][1]
        data['book_rating'] = book_rating

        master_list.append(data)
    
    print(f"Page {page} scrapped successfully!\n")
    
end_clock = time.process_time_ns()
print(f"Total Scraping time: {end_clock - start_clock} ns")
```
